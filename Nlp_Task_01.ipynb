# **Task 1: Conceptual Understanding**

**1. What is the difference between "Love" and "love" in NLP?**\
**Ans:** In NLP, text is processed as tokens, and by default, it is case-sensitive. This means "Love" and "love" are treated as two separate tokens with independent meanings and counts. "Love" may appear at the beginning of sentences or as a proper noun, whereas "love" is commonly used as a verb or noun. Without normalization techniques like lowercasing, this distinction increases vocabulary size unnecessarily and reduces the model’s ability to generalize patterns effectively.

**2. What happens if stopwords are not removed?**\
**Ans:** If stopwords are retained, they tend to appear frequently across almost all documents, making them less useful for distinguishing between texts. Their presence can dilute the importance of meaningful terms, leading to weaker feature representation. Additionally, models may require more processing time and memory, as they need to handle a larger number of irrelevant words that do not contribute significantly to understanding the content.

**3. Provide two real-world scenarios where removing stopwords can be harmful.\
Ans:** 
*   **Legal Documents Analysis:**
In legal texts, words like "shall", "may", or "not" are critical for interpreting obligations and permissions. Removing them can completely change the meaning of clauses and lead to incorrect interpretations.
*   **Voice Assistants / Chatbots:**
In conversational queries like "Can you not play music?", removing "not" would change the intent to "play music", causing the system to perform the opposite action of what the user intended.
*   **Sentiment Analysis:** In the sentence "I am not happy," the word "not" is a stopword. If you remove it, the sentence becomes "I happy," which completely flips the sentiment from negative to positive.
*   **Search Engines / Intent Recognition:** For a query like "To be or not to be," every single word is technically a stopword. Removing them would leave the search engine with zero input, losing the entire context of the famous quote.

**4. Explain the difference between stemming and lemmatization.\
Ans:**

*   **Stemming:** This is a "crude" process that chops off the ends of words to find the root. It often follows fixed rules, which can lead to non-dictionary words. For example, "studies" and "studying" might both become "studi". It is fast but less accurate.Stemming reduces words to their base form by removing prefixes or suffixes using simple heuristic rules. It does not consider the context or grammar, so the output may not always be a valid word (e.g., "running" → "run", but "happiness" → "happi"). It is faster and commonly used in large-scale text processing.
*   **Lemmatization:** This is a more sophisticated process that uses a dictionary (morphological analysis) to return the word to its actual base form, known as a "Lemma." For example, "better" would be lemmatized to "good." It is more computationally expensive but highly accurate.Lemmatization converts words to their meaningful base form by considering vocabulary and grammatical structure. It ensures that the result is a valid word (e.g., "better" → "good", "running" → "run"). Though slower than stemming, it provides more accurate and linguistically meaningful results.

# **Task 2:Build Advanced Preprocessing Function**

Implement the following function:
def preprocess_text(text):
    pass

Required Features

Your function must include:

Remove numbers
Example: "I have 2 dogs" → "I have dogs"

Remove extra spaces
Example: "This is good" → "This is good"

Handle repeated characters
Example: "soooo goooood!!!" → "so good"

Convert text to lowercase
Example: "WOW!!! This is GREAT!!!" → "wow this is great"

Remove very short tokens
Remove words with length ≤ 2
Exception: keep meaningful words like "no", "not"

Remove URLs and email-like patterns
Example: "Visit http://example.com now" → "visit now"

```python
import re

def preprocess_text(text):

    if not isinstance(text, str):
        return ""

    # 1. Remove numbers
    # Matches any digit and replaces it with nothing
    text = re.sub(r'\d+', '', text)

    # 2. Remove extra spaces
    # Replaces multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()

    # 3. Handle repeated characters
    # Reduces characters appearing 3 or more times to just 2 (e.g., "soooo" -> "so")
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # 4. Convert text to lowercase
    # Ensures uniform analysis by ignoring capitalization
    text = text.lower()

    # 5. Remove very short tokens (Length ≤ 2)
    # Exception: keep meaningful words like "no" and "not"
    meaningful_exceptions = {'no', 'not'}
    tokens = text.split()
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful_exceptions]
    text = " ".join(tokens)

    # 6. Remove URLs and email-like patterns
    # Matches web protocols (http/https), www, and common email structures
    text = re.sub(r'https?://\S+|www\.\S+|\S+@\S+', '', text)

    # Final cleanup of punctuation (for clean tokens)
    text = re.sub(r'[^\w\s]', '', text).strip()

    return text

# TEST SUITE (Verifying based on examples)
print(f"1. Numbers:          {preprocess_text('I have 2 dogs')}")
print(f"2. Extra Spaces:     {preprocess_text('This   is     good')}")
print(f"3. Repeated Chars:   {preprocess_text('soooo goooood!!!')}")
print(f"4. Lowercase:        {preprocess_text('WOW!!! This is GREAT!!!')}")
print(f"5. Short/Exceptions: {preprocess_text('he is no not happy')}")
print(f"6. Web Patterns:     {preprocess_text('Visit http://example.com now')}")
```

# **Task 4: Token Analytics**

For each processed sentence, compute:

Total number of tokens

Number of unique tokens

Average token length

Analysis Questions

Which sentence had the most noise?

Which sentence retained the most meaningful tokens after cleaning?

```python
import re

texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# preprocessing function
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)

    words = text.split()
    stopwords = ["i", "am", "is", "the", "this", "at", "me", "now", "with"]

    result = []
    for w in words:
        if w not in stopwords:
            w = re.sub(r"(.)\1+", r"\1", w)
            result.append(w)

    return result


max_noise = ""
max_noise_count = 0

max_meaningful = ""
max_meaningful_count = 0

print("Token Analytics \n")

for i, text in enumerate(texts, 1):
    tokens = clean_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_len = sum(len(t) for t in tokens) / total_tokens
    else:
        avg_len = 0

    print(f"Sentence {i}")
    print("Tokens:", tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_len, 2))
    print("-" * 40)

    # track most noise (original length - cleaned tokens)
    noise = len(text.split()) - total_tokens
    if noise > max_noise_count:
        max_noise_count = noise
        max_noise = text

    # track most meaningful (max tokens after cleaning)
    if total_tokens > max_meaningful_count:
        max_meaningful_count = total_tokens
        max_meaningful = text


print("\nMost Noise Sentence:")
print(max_noise)

print("\nMost Meaningful Sentence:")
print(max_meaningful)
```

# **Task 5: Frequency Analysis**

Combine all tokens and use:

Identify:

```python
import re
from collections import Counter

texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# simple preprocessing
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)

    words = text.split()
    stopwords = ["i", "am", "is", "the", "this", "at", "me", "now", "with"]

    result = []
    for w in words:
        if w not in stopwords:
            w = re.sub(r"(.)\1+", r"\1", w)
            result.append(w)

    return result

# combine all tokens
all_tokens = []

for text in texts:
    all_tokens.extend(clean_text(text))

# frequency count
freq = Counter(all_tokens)

# top 10 most frequent
print("Top 10 Most Frequent Words:")
print(freq.most_common(10))

# least frequent (count = 1)
least = [word for word, count in freq.items() if count == 1]

print("\nTop 5 Least Frequent Words:")
print(least[:5])
```
